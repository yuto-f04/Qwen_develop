## 実行手順
1. Cell 1: リポジトリの取得 (初回は clone、2回目以降は pull)
2. Cell 2: 依存パッケージのインストール
3. Cell 3: Gradio アプリ起動 (share=True で公開 URL を取得)

```python
# ===== Cell 1: リポジトリの取得 =====
import os

REPO_URL = "https://github.com/yuto-f04/Qwen_develop.git"
REPO_PATH = "/content/Qwen_develop"

if os.path.exists(REPO_PATH):
    print("リポジトリ取得済み。最新版に更新します...")
    !git -C {REPO_PATH} pull
else:
    print("リポジトリをクローンします...")
    !git -C /content clone {REPO_URL}

# 絶対パスで移動することで何度実行しても入れ子にならない
%cd {REPO_PATH}
print(f"作業ディレクトリ: {os.getcwd()}")
```

```python
# ===== Cell 2: 依存パッケージのインストール =====
# torch は Colab に事前インストール済みのため除外
# sox: qwen_tts が音声デコードに使う OS バイナリ（pip では入らない）
!apt-get install -y sox libsox-fmt-all
!pip install -q yt-dlp
!pip install -q -r requirements.txt
# flash-attn: 推論を高速化（なくても動くが数倍遅くなる）
# ninja・packaging を先に入れないとビルドが失敗する
!pip install -q packaging ninja
!pip install flash-attn --no-build-isolation
```

```python
# ===== Cell 3: Gradio アプリ起動 =====
# share=True で公開 URL が発行される (約72時間有効)
!python app.py
```
